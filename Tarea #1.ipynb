# ¿Qué son los portafolios de inversión?

Bueno segun la revista Forbes.(2022) un portafolio de inersión es el conjunto total de activos financieros que una persona tiene, puede estar compuesto por instrumentos de renta fija, variable o mixta como bonos, acciones, monedas, efectivo, materias primas y productos derivados.
