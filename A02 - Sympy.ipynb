$f(x)=1260+117x-9x^2$

```python
import sympy as sp
import numpy as np
import pandas as pd

sp.var("x")
def f(x):
    return  1260 + 117*x - 9*(x)**2
f(x)
```

```python
df= sp.diff(f(x), x)
df
```

```python
pc= sp.solve(df, x)
pc
```

```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline

xn= np.linspace(-20,30,1000)
yn= f(xn)


plt.figure()
plt.plot(xn,yn)
plt.grid()
plt.plot(pc[0], f(pc[0]), "*r")
```
